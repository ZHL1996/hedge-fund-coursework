```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = 'Hedge_Fund/data/PS_6_Industry_Momentum (1).xlsx'
factor = pd.read_excel(data,sheet_name='factors',index_col=0,parse_dates=['date'])
```

```python
ind=pd.read_excel(data,sheet_name='industry_portfolios',
                  index_col='date',parse_dates=['date'])
ind.head()
```

```python
ind2= (ind.rolling(12).mean()).shift().dropna()
ind2
```

```python
rank = ind2.rank(axis=1,ascending=True)
rank
```

```python
rank.mean(axis=0).sort_values() # lowest:Telcm, highest:ElcEq
```

```python
print('minimun rank is:',rank.mean(axis=0).sort_values().idxmin()+ str(rank.mean(axis=0).sort_values().min())) 
print('max rank is:',rank.mean(axis=0).sort_values().idxmax()+ str(rank.mean(axis=0).sort_values().max())) 
```

```python
rank['Autos'].plot(figsize=(10,6),title='Autos rank over time')
plt.xlabel('time')
plt.ylabel('ranks')
plt.show();
```

```python
index = rank.mean(axis=0).nlargest(5).index
rank[index].plot(subplots=True,figsize=(10,6),title='ranks over time');
```

```python
pd.Series(rank.idxmax(axis=1).value_counts()).plot.bar(figsize=(9,6),title='top rank stocks');
```

```python
wp = ((rank>15) * ind).dropna()
wp
```

```python
wp_rets = (wp.sum(axis=1))/15
wp_rets = (wp_rets - factor['RF']).dropna()
wp_rets
```

```python
def fun(x):
    mean = x.mean()
    std = x.std()
    sr = mean/std
    a_mean = mean*12
    a_std = std*np.sqrt(12)
    a_sr = sr * np.sqrt(12)
    avg_mean,mon_mean = str(round(a_mean*100,2)) + "%",str(round(mean*100,2)) + "%"
    ann_std,mon_std = str(round(a_std*100,2))+"%", str(round(std*100,2))+"%"
    return pd.DataFrame(data=[[mon_mean,mon_std,round(sr,2)],[avg_mean,ann_std,round(a_sr,2)]],
                        columns=['rets','std','sr'],
                       index=['monthly','annually'])
print("Winners:")
fun(wp_rets)
```

```python
lp = (rank<=15)
lp_rets = ((((lp*ind).dropna()).sum(axis=1))/15 - factor['RF']).dropna()
lp_rets
```

```python
print('Loser:')
fun(lp_rets)
```

```python
print('Market:')
fun((factor['Mkt-RF']).reindex(lp_rets.index))
```

```python
indmom = wp_rets - lp_rets
indmom
```

```python
fun(indmom)
```

```python
y = indmom
X = factor['Mkt-RF'].reindex(y.index)
```

```python
import statsmodels.api as sm
```

```python
def fun2(n):
    mask1 = rank > (30-n)
    rets1 = (((mask1 * ind).sum(axis=1)).dropna())/(n)
    mask2 = rank <= n
    rets2 = (((mask2 * ind).sum(axis=1)).dropna())/(n)
    ind_mom = rets1 - rets2 
    #rf = (factor['RF'].reindex(rets1/index)).mean()
    mean = (ind_mom).mean() #- rf
    std = (ind_mom).std()
    sr = mean / std
    return sr
```

```python
fun2(19)
```

```python
range = np.linspace(1,29,29)
sr = {}
for i in range:
    sr.update({i:fun2(i)})   
x = pd.DataFrame(data=sr.values(),index = sr.keys())
x.plot(xlabel='size',ylabel = 'sr',title='optimization size',
       legend=False,figsize=(10,6),grid=True);
```
