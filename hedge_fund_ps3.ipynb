```python
import pandas as pd
import numpy as np
path = 'Hedge_Fund/'
data = path + "PS_3_Exercise_XS_Regressions (2).xlsx"
data
```

```python
xrets = pd.read_excel(data, sheet_name='excess_returns',index_col='year')
xrets
```

```python
s1 = pd.read_excel(data, sheet_name='signal_1',index_col='year')
s1
```

```python
s2 = pd.read_excel(data, sheet_name='signal_2',index_col='year')
s2
```

```python
import statsmodels.api as sm
```

```python
xrets.iloc[1,:]
```

```python
alpha = []
beta = []
for i in range(1,xrets.shape[0]):
    y = xrets.iloc[i,:]
    x = s1.iloc[i-1,:]
    x=sm.add_constant(x)
    reg = sm.OLS(y,x).fit()
    alpha.append(reg.params.iloc[0])
    beta.append(reg.params.iloc[1])
```

```python
ind = list(xrets.index)
ind.remove(1999)
ind
```

```python
uv_s1 = pd.Series(data=beta,index = ind)
```

```python
alpha1 = []
beta1 = []
for i in range(1,xrets.shape[0]):
    y = xrets.iloc[i,:]
    x = s2.iloc[i-1,:]
    x=sm.add_constant(x)
    reg1 = sm.OLS(y,x).fit()
    alpha1.append(reg1.params.iloc[0])
    beta1.append(reg1.params.iloc[1])
```

```python
uv_s2 = pd.Series(data=beta1,index = ind)
```

```python
alpha2 = {}
beta2 = {}
for i in range(1,xrets.shape[0]):
    y = xrets.iloc[i,:]
    x = [s1.iloc[i-1,:],s2.iloc[i-1,:]]
    x=pd.concat(x,axis=1)
    x.columns=['s1','s2']
    x=sm.add_constant(x)
    reg3 = sm.OLS(y,x).fit()
    ind = xrets.index[i]
    alpha2.update({ind:reg3.params.iloc[0]})
    beta2.update({ind:[reg3.params.iloc[1],reg3.params.iloc[2]]})
mv = pd.DataFrame(beta2,index=['s1','s2']).T
mv
```

```python
print('mv:')
print("average signal 1 is:",round(mv.T.loc['s1'].mean(),2))
print("average signal 2 is:",round(mv.T.loc['s2'].mean(),2))
```

```python
uv = pd.concat([uv_s1,uv_s2],axis=1)
uv.columns = ['s1','s2']
uv
```

```python
uv.plot.hist();
mv.plot.hist();
```

```python
beta = {}
for i in range(xrets.shape[1]):
    y = xrets.drop(1999).iloc[:,i]
    x = [s1.shift(1).dropna().iloc[:,i],s2.shift(1).dropna().iloc[:,i]]
    x=pd.concat(x,axis=1)
    x = sm.add_constant(x)
    index = xrets.columns[i]
    reg = sm.OLS(y,x).fit()
    beta.update({index:[reg.params[1],reg.params[2]]})
ts = pd.DataFrame(beta,index = ['s1','s2'])
ts
```

```python
print('time series mean:')
print(ts.mean(axis=1))
```
